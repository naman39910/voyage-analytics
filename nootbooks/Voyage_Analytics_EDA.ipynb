## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# convert date column into datetime format
flights['date'] = pd.to_datetime(flights['date'], format='%m/%d/%Y')
hotels['date'] = pd.to_datetime(hotels['date'], format='%m/%d/%Y')

# Extract month and year from the date column
flights['months'] = flights['date'].dt.month
flights['year'] = flights['date'].dt.year
hotels['months'] = hotels['date'].dt.month
hotels['year'] = hotels['date'].dt.year
flights.head()
````

````python
hotels.head()
````

````python
# Function to get price statistics per flight type
def get_price_stats(df, category_col):
    '''
    This function returns mean and median price for each category.
    '''
    try:
        return df.groupby(category_col)['price'].agg(['mean','median','min','max']).round(2)
    except Exception as e:
        print(e)

print("===== Price Stats by Flight Type =====")
get_price_stats(flights, 'flightType')
````

````python
# Price stats by agency
print("===== Price Stats by Agency =====")
get_price_stats(flights, 'agency')
````

````python
# Top 10 Most Expensive Routes
top_routes = flights.groupby(['from', 'to'])['price'].mean().sort_values(ascending=False).head(10).round(2)
print(top_routes)
````

````python
# Most booked hotels
print("===== Top 5 Most Booked Hotels =====")
print(hotels['name'].value_counts().head())
````

````python
# user age segmentations
def age_segment(age):
    if age < 30:
        return 'Young'
    elif age < 50:
        return 'Middle-aged'
    else:
        return 'Old'

users['age_segment'] = users['age'].apply(age_segment)
print(users['age_segment'].value_counts())
````
